```python
from utils import make_w2v_embeddings, load_word2vec, load_data, split_and_zero_padding

import numpy as np
```

```python
def save_data(file_path, obj):
  with open(file_path, "wb") as f:
    np.save(f, obj)
```

### Load Word2Vec and Data

```python
embedding_dict = load_word2vec("./GoogleNews-vectors-negative300.bin.gz")
```

```python
data_quora = load_data("./quora_train_10000.csv")
```

### Create embedding

```python
train_df, embeddings, vocabs = make_w2v_embeddings(embedding_dict, data_quora, embedding_dim=300)
```

### Split train test dataset

```python
msk = np.random.rand(len(train_df)) < 0.85
```

```python
train = train_df[msk]
```

```python
test = train_df[~msk]
```

Dataset

```python
train.to_csv("./data/train.csv")
test.to_csv("./data/test.csv")
```

```python
X = train[['question1_n', 'question2_n']]
Y = train['is_duplicate']
```

```python
X_test = test[['question1_n', 'question2_n']]
Y_test = test['is_duplicate']
```

```python
X = split_and_zero_padding(X, 30)
# X_dev = split_and_zero_padding(X_dev, 30)
X_test = split_and_zero_padding(X_test, 30)
```

```python
save_data("./data/numpy/train/sentence_1.npy", X["left"])
save_data("./data/numpy/train/sentence_2.npy", X["right"])
save_data("./data/numpy/train/label.npy", Y.values)
# save_data("./data/dev/sentence_1.npy", X_dev["left"])
# save_data("./data/dev/sentence_2.npy", X_dev["right"])
# save_data("./data/dev/label.npy", Y_dev.values)
save_data("./data/numpy/test/sentence_1.npy", X_test["left"])
save_data("./data/numpy/test/sentence_2.npy", X_test["right"])
save_data("./data/numpy/test/label.npy", Y_test.values)
```
